Data cleaning and Data preparation:
    
    1. Take a copy of the dataset and work on that. Any changes made on the copy will not affect the main dataset.
    2. Check for duplicate values.
    3. check for null values.
    4. Replacing null values.
    5. Remove unwanted columns.
    6. Change the column names if you want.
    7. Add few columns for further analysis if you want.

```python
#  Taking a copy of the dataset

df_copy = df.copy()
```

```python
df_copy.shape
```

```python
df_copy.info()
```

```python
# Dropping unwanted columns

df_copy.drop('Unnamed: 0', axis =1, inplace = True)
```

```python
df_copy.columns
```

```python
# Checking how many null values in each column in a dataset 

df_copy.isnull().sum()
```

```python
df_copy[df_copy['Cost IN $ (000)'].isnull()]
```

*   These players were went unsold in 2022 auction, so their cost will replace with zero.

```python
# Replacing na values with ZERO

df_copy['Cost IN $ (000)'] = df_copy['Cost IN $ (000)'].fillna(0)
df_copy['COST IN ₹ (CR.)'] = df_copy['COST IN ₹ (CR.)'].fillna(0)                               
```

```python
df_copy.head()
```

```python
df_copy['2021 Squad'] = df_copy['2021 Squad'].fillna('Not Participated in IPL 2021')
df_copy
```

```python
df_copy.isnull().sum()
```

*  Here, All null values are replaced.

```python
# Adding few columns for further analysis
```

```python
teams = df_copy[df_copy['COST IN ₹ (CR.)']>0]['Team'].unique()
teams
```

```python
df_copy['status'] = df_copy['Team'].replace(teams,'sold')
```

```python
df_copy['Base Price'].unique()
```

```python
df_copy['retention'] = df_copy['Base Price']
```

```python
df_copy.head()
```

```python
# Analyzing the Base Price column 
```

```python
# Replacing Draft Pick with zero

df_copy['Base Price'].replace('Draft Pick',0, inplace = True)

df_copy.head(5)
```

```python
# The Base Price column splitting into two new columns: "base_price_unit" and "base_price".

# 1. The 'base_price_unit' column will contain the last word of the Base Price column,which is the unit of the base price.
# 2. The 'base_price' column will contain the first word of the Base Price column,which is the base price without the unit.

df_copy['base_price_unit'] = df_copy['Base Price'].apply(lambda x: str(x).split(' ')[-1])   
df_copy['base_price'] = df_copy['Base Price'].apply(lambda x: str(x).split(' ')[0])
```

```python
# Replacing Retained with zero, because those players are NOT Participated in Auction,
# the teams are retain some players before the auction.

df_copy['base_price'].replace('Retained',0,inplace=True)
```

```python
df_copy['base_price_unit'].unique()
```

```python
df_copy['base_price_unit'] = df_copy['base_price_unit'].replace({'Cr':100,'Lakh':1,'Retained':0})
```

```python
df_copy['base_price'] = df_copy['base_price'].astype(float)
df_copy['base_price_unit'] = df_copy['base_price_unit'].astype(int)
```

```python
df_copy['base_price'] = df_copy['base_price']*df_copy['base_price_unit']
```

```python
df_copy.head(5)
```

```python
# Removing unwanted columns

df_copy.drop(['Base Price','base_price_unit'], axis =1, inplace = True)
```

```python
df_copy.head(2)
```

```python
df_copy['COST IN ₹ (CR.)'] = df_copy['COST IN ₹ (CR.)']*100
```

```python
df_copy = df_copy.rename(columns={'TYPE':'Type','COST IN ₹ (CR.)':'Sold_for_lakhs',
                        'Cost IN $ (000)':'Cost_in_dollars',
                        '2021 Squad':'Prev_team','Team':'Curr_team'})
```

```python
df_copy.head(2)
```

```python
# Checking for duplicates

df_copy[df_copy['Player'].duplicated(keep=False)]
```

* There are duplicate rows, but the name of the players is same, but there are differnt players.

# 2. Participation based on the Role(Batsman, Bowlers, Allrounders and        WicketKeeper)

```python
type = df_copy['Type'].value_counts()
type.reset_index()
```

```python

# Create a pie chart of the number of players in each role, with labels and percentages
plt.pie(type.values,labels=type.index,labeldistance=1.1,autopct='%1.2f%%')

# Set the title of the plot
plt.title("DIFFERENT ROLES PARTICIPATED IN IPL 2022 AUCTION")

# Display the plot
plt.show()
```

# 3. How many Players Were Sold in IPL 2022 Auctions ?

```python
sold_or_unsold =df_copy['status'].value_counts()
sold_or_unsold.reset_index()
```

```python
# Create a list of colors
c=['red','green']

# Create a bar plot of the number of players sold and unsold, with labels and values
sold_or_unsold.plot.bar(x=sold_or_unsold.index, y=sold_or_unsold.values,width=0.5,color=c)
plt.xticks()

# Set the title and labels
plt.title('Sold vs Unsold')
plt.xlabel('Sold or Unsold')
plt.ylabel('Number of Players')


# Add labels to the bars
for p in plt.gca().patches:
    plt.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width()/2, p.get_height()),
                 ha = 'center', va = 'center', xytext = (0, 4), textcoords = 'offset points')

# Display the plot   
plt.show()
```

*  UNSOLD = 396
*    SOLD = 237 
*  396 players were unsold and 237 players were sold.
